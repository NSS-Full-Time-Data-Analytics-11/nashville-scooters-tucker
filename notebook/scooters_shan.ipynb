```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sqlalchemy import create_engine, text
```

```python
database_name = 'scooters'    

connection_string = f"postgresql://postgres:postgres@localhost:5432/{database_name}"
```

```python
engine = create_engine(connection_string)
```

```python
query = '''
SELECT *
FROM scooters
LIMIT 200;
'''

with engine.connect() as connection: result = connection.execute(text(query))
```

```python
with engine.connect() as connection:    
    scooters = pd.read_sql(text(query), con = connection)
```

```python
scooters.head()
```

```python
query = '''
SELECT EXTRACT(MONTH from DATE(pubdatetime)) AS month, pubdatetime, companyname, sumdid, latitude, longitude, chargelevel, costpermin
FROM scooters
WHERE EXTRACT(MONTH from DATE(pubdatetime)) = 5 AND sumdgroup <> 'bicycle' AND companyname = 'Lyft';
'''
with engine.connect() as connection: result = connection.execute(text(query))
with engine.connect() as connection:
    lyft_may_scooters = pd.read_sql(text(query), con = connection)
lyft_may_scooters.head()
```
